```python
from sklearn import mixture
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.metrics.cluster import contingency_matrix
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("data_final_linear.csv")
X = df.iloc[:,0:14].values
df
```

```python
Y = []
for i in df['xAttack']:
    if i=="normal":
        Y.append(0)
    elif i=="dos":
        Y.append(1)
    elif i=="r2l":
        Y.append(2)
    elif i=="probe":
        Y.append(3)
    elif i=="u2r":
        Y.append(4)
```

```python
gmm = mixture.GaussianMixture(n_components=5).fit(X)
```

```python
labels = gmm.predict(X)
```

```python
cn = contingency_matrix(labels,Y)
ans = 0
for i in cn:
    ans = ans + max(i)
purity = ans/len(df)
purity
```
